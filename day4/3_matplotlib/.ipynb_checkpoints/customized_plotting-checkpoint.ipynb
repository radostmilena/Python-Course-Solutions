# Data visualistation with matplotlib


```python
import matplotlib
import matplotlib.pyplot as plt
```

This is a pie chart of randomly generated data, labeled with a single letter from a randomly generated string (function found here: https://pynative.com/python-generate-random-string/). It also 'explodes' the last slice of data, meaning it will be slightly shifted outwards.

```python
import numpy as np
from matplotlib.cm import get_cmap
import string
import random

def get_random_string(length):
    letters = string.ascii_lowercase
    result_str = ''.join(random.choice(letters) for i in range(length))
    return(result_str)

name = "tab20b"
cmap = get_cmap(name) 
colors = cmap.colors  

data = np.random.rand(15)

explode = np.zeros(15)
explode[-1] = 0.2

string = get_random_string(15)
labels = list(string)
print(labels)
```

In plotting, I also set the shadow to be on and the labeling of slices to be formatted as percentages.

```python
fig = plt.figure()
ax  = fig.add_subplot(111)

ax.pie(data, explode = explode, labels = labels, colors = colors, shadow = True, autopct='%.0f%%')

plt.show()
fig.savefig('my_fancy_plot.pdf', format='pdf', bbox_inches='tight')
```
